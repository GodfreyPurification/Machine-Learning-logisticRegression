```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/MachineLearning/assignment/dataset/bankloan.csv')
df=df.drop(columns=['ID','ZIP.Code'],axis=1) 
df
```

```python
df.head()
```

```python
X = df.drop(columns=["Personal.Loan"])  # Features
y = df["Personal.Loan"]  # Target variable
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_x=LabelEncoder()
x = X.apply(le_x.fit_transform)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.30, random_state=1)
```

```python
xtrain
```

```python
xtest
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dect = DecisionTreeClassifier()
```

```python
dect.fit(xtrain,ytrain)
```

```python
dect.score(xtest, ytest)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
result = cross_val_score(DecisionTreeClassifier(), X, y,cv=5)
result
```

```python
result.mean()
```

```python
from sklearn.linear_model import LogisticRegression
result = cross_val_score(LogisticRegression(), X, y,cv=5)
result
```

```python
result.mean()
```

```python
y_predict = dect.predict(xtest)
y_predict
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(ytest,y_predict))
```
